```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
import numpy as np
import os

#read csv
conv = pd.read_csv('CO2 Emissions_Canada.csv')
bev  = pd.read_csv('my2012-2026-battery-electric-vehicles.csv')
phev = pd.read_csv('my2012-2026-plug-in-hybrid-electric-vehicles.csv')

print(f'conventional/hybrid: {len(conv)} rows')
print(f'BEV：{len(bev)} rows')
print(f'PHEV：{len(phev)} rows')

# integrate csv
def extract_le(s):
    try: return float(str(s).split()[0])
    except: return np.nan

conv_out = pd.DataFrame({
    'Model_Year':               np.nan,
    'Make':                     conv['Make'],
    'Model':                    conv['Model'],
    'Vehicle_Class':            conv['Vehicle Class'].str.title(),
    'Vehicle_Type':             'Conventional/Hybrid',
    'Engine_Size_L':            conv['Engine Size(L)'],
    'Cylinders':                conv['Cylinders'],
    'Motor_kW':                 np.nan,
    'Transmission':             conv['Transmission'],
    'Fuel_Type_Primary':        conv['Fuel Type'],
    'Fuel_Type_Secondary':      np.nan,
    'Fuel_Cons_City_L100km':    conv['Fuel Consumption City (L/100 km)'],
    'Fuel_Cons_Hwy_L100km':     conv['Fuel Consumption Hwy (L/100 km)'],
    'Fuel_Cons_Comb_L100km':    conv['Fuel Consumption Comb (L/100 km)'],
    'Fuel_Cons_Comb_mpg':       conv['Fuel Consumption Comb (mpg)'],
    'Elec_Cons_City_kWh100km':  np.nan,
    'Elec_Cons_Hwy_kWh100km':   np.nan,
    'Elec_Cons_Comb_kWh100km':  np.nan,
    'Elec_Range_km':            np.nan,
    'Total_Range_km':           np.nan,
    'Recharge_Time_h':          np.nan,
    'CO2_Emissions_g_km':       conv['CO2 Emissions(g/km)'],
    'CO2_Rating':               np.nan,
    'Smog_Rating':              np.nan,
})

bev_out = pd.DataFrame({
    'Model_Year':               bev['Model year'],
    'Make':                     bev['Make'],
    'Model':                    bev['Model'],
    'Vehicle_Class':            bev['Vehicle class'],
    'Vehicle_Type':             'Battery Electric (BEV)',
    'Engine_Size_L':            np.nan,
    'Cylinders':                np.nan,
    'Motor_kW':                 bev['Motor (kW)'],
    'Transmission':             bev['Transmission'],
    'Fuel_Type_Primary':        bev['Fuel type'],
    'Fuel_Type_Secondary':      np.nan,
    'Fuel_Cons_City_L100km':    np.nan,
    'Fuel_Cons_Hwy_L100km':     np.nan,
    'Fuel_Cons_Comb_L100km':    np.nan,
    'Fuel_Cons_Comb_mpg':       np.nan,
    'Elec_Cons_City_kWh100km':  bev['City (kWh/100 km)'],
    'Elec_Cons_Hwy_kWh100km':   bev['Highway (kWh/100 km)'],
    'Elec_Cons_Comb_kWh100km':  bev['Combined (kWh/100 km)'],
    'Elec_Range_km':            bev['Range (km)'],
    'Total_Range_km':           bev['Range (km)'],
    'Recharge_Time_h':          bev['Recharge time (h)'],
    'CO2_Emissions_g_km':       bev['CO2 emissions (g/km)'],
    'CO2_Rating':               bev['CO2 rating '].replace('n/a', np.nan),
    'Smog_Rating':              bev['Smog rating'].replace('n/a', np.nan),
})

phev_out = pd.DataFrame({
    'Model_Year':               phev['Model year'],
    'Make':                     phev['Make'],
    'Model':                    phev['Model'],
    'Vehicle_Class':            phev['Vehicle class'],
    'Vehicle_Type':             'Plug-in Hybrid (PHEV)',
    'Engine_Size_L':            phev['Engine size (L)'],
    'Cylinders':                phev['Cylinders'],
    'Motor_kW':                 phev['Motor (kW)'],
    'Transmission':             phev['Transmission'],
    'Fuel_Type_Primary':        phev['Fuel type 1'],
    'Fuel_Type_Secondary':      phev['Fuel type 2'],
    'Fuel_Cons_City_L100km':    phev['City (L/100 km)'],
    'Fuel_Cons_Hwy_L100km':     phev['Highway (L/100 km)'],
    'Fuel_Cons_Comb_L100km':    phev['Combined (L/100 km)'],
    'Fuel_Cons_Comb_mpg':       np.nan,
    'Elec_Cons_City_kWh100km':  np.nan,
    'Elec_Cons_Hwy_kWh100km':   np.nan,
    'Elec_Cons_Comb_kWh100km':  phev['Combined Le/100 km'].apply(extract_le),
    'Elec_Range_km':            phev['Range 1 (km)'],
    'Total_Range_km':           phev['Range 1 (km)'].fillna(0) + phev['Range 2 (km)'].fillna(0),
    'Recharge_Time_h':          phev['Recharge time (h)'],
    'CO2_Emissions_g_km':       phev['CO2 emissions (g/km)'],
    'CO2_Rating':               phev['CO2 rating'].replace('n/a', np.nan),
    'Smog_Rating':              phev['Smog rating'].replace('n/a', np.nan),
})

df = pd.concat([conv_out, bev_out, phev_out], ignore_index=True)
print(f'\nfinish integration, total {len(df)} rows，{df.shape[1]} fields')

#save integrated csv
df.to_csv('vehicles_integrated.csv', index=False)
print('\n✅ data saved：vehicles_integrated.csv')
```
